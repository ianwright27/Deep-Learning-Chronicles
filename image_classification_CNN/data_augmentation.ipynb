### **Data Augmentation with Tensorflow's Keras API**

Just an additional log. 

Data Augmentation occurs when we create new data by making modifications on already existing data. 

- Specifically for Image data (since we have been working with it), that would entail things such as:
    - flipping (horizontally / vertically)
    - rotating 
    - changing the color of the image 
    - etc

- Why would we need this?
Well... to get access to more data, or simply:
    - **to grow our training set** 

Alot of times, not having access to enough data, is an issue that we can run into, and that may result into problems like **overfitting** when our training data is too small. 

Importing the necessary packages

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
# This function will plot images in the form of a grid with 1 row and 10 columns
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20, 20))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes): 
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
gen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1, height_shift_range=0.1, shear_range=0.15, zoom_range=0.1, channel_shift_range=10., horizontal_flip=True)
```

```python

# os.chdir('Documents/Deep-Learning-Chronicles/')
# os.chdir('image_classification_CNN')
os.getcwd()
os.listdir()
```

```python
chosen_image = random.choice(os.listdir('mobilenet_data/data'))
```

```python
image_path = 'mobilenet_data/data/'+chosen_image
```

```python
assert os.path.isfile(image_path)
```

```python
# obtain image 
image = np.expand_dims(plt.imread(image_path),0)
plt.imshow(image[0])
```

```python
# Generate batches of augmented images from this image
aug_iter = gen.flow(image)
```

```python
# Get 10 samples of augmented images 
aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
# Augmented images
plotImages(aug_images)
```

10 variations of the same image have now been generated according to the parameters of the ImageDataGenerator, using **data augmentation**. 

That's the whole point of growing our training set.
